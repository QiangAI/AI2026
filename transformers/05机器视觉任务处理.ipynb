# 1. Transformers的机器视觉任务概述

- 在Transformers框架中，把机器视觉的任务分成如下几个主题：
    - Depth Estimation(279 models)
        - 深度估计是一项预测图像中物体深度信息的任务。
    - Image Classification(25,474 models)
        - 图像分类是为整张图像分配一个标签或类别的任务。每张图像通常只对应一个类别。该模型以图像作为输入，并输出该图像所属类别的预测结果。
    - Image Feature Extraction(1,003 models)
        - 图像特征提取是从计算机视觉模型中提取已学习特征的任务。
    - Image Segmentation(2,204 models)
        - 图像分割是将图像划分为多个区域的任务，其中图像中的每个像素都被映射至特定的物体。此任务包含多种类型，例如实例分割、全景分割和语义分割。
    - Image-to-Image(2,076 models)
        - 图像到图像转换是指通过超分辨率、图像修复、着色等多种可能的处理与增强技术，对输入图像进行转换的任务。
    - Image-to-Text(1,304 models)
        - 图像到文本模型能够根据给定的图像输出相应的文本描述。图像描述生成和光学字符识别可被视为图像到文本领域最常见的应用。
    - Image-to-Video(591 models)
        - 图像到视频模型以静态图像作为输入，并生成一段视频。这些模型可以通过文本提示进行引导，以影响输出视频的内容和风格。
    - Keypoint Detection(226 models)
        - 关键点检测是指识别图像中具有显著特征或代表性位置点的任务。
    - Mask Generation(185 models)
        - 掩码生成是指为给定图像生成特定目标或感兴趣区域标识掩码的任务。这类掩码常用于分割任务，通过精准分离目标对象以支持后续处理或分析。
    - Object Detection(4,428 models)
        - 目标检测模型能够帮助用户识别图像中属于特定类别的物体。这类模型以图像作为输入，输出带有边界框和类别标签的检测结果。
    - Video Classification(2,036 models)
        - 视频分类是为整个视频分配一个标签或类别的任务。每个视频通常只对应一个类别。视频分类模型以视频作为输入，并输出该视频所属类别的预测结果。
    - Text-to-Image(91,800 models)
        - 文本到图像是指根据输入的文本生成图像的任务。这类流程同样可以基于文本提示对图像进行修改和编辑。
    - Text-to-Video(962 models)
        - 文本到视频模型可应用于任何需要根据文本生成连贯图像序列的场景。
    - Unconditional Image Generation(2,443 models)
        - 无条件图像生成是指在不依赖任何上下文条件（如提示文本或另一张图像）的情况下生成图像的任务。模型完成训练后，将能够生成符合其训练数据分布的图像。
    - Video-to-Video(89 models)
        - 视频到视频模型以一个或多个视频作为输入，并生成新的视频作为输出。这些模型能够提升画质、进行帧插值、调整风格或创建新的动态效果，从而支持创意应用、视频制作及学术研究。
    - Zero-Shot Image Classification(1,126 models)
        - 零样本图像分类是指在模型训练阶段对未见过的新类别进行分类的任务。
    - Zero-Shot Object Detection(100 models)
        - 零样本目标检测是一项计算机视觉任务，旨在无需预先训练或了解特定类别的情况下，检测图像中的物体及其类别。零样本目标检测模型以图像和候选类别列表作为输入，输出检测到的物体边界框和标签。
    - Text-to-3D(103 models)
        - 文本到3D模型是指根据输入的文本生成三维模型的任务。
    - Image-to-3D(408 models)
        - 图像到3D模型是指根据输入的图像生成三维模型的任务。

# 2. DepthEstimationPipeline与Depth Estimation任务

- 深度估计模型可用于估算图像中不同物体的深度信息。主要常见应用场景包含：
    - 体积信息估算（Estimation of Volumetric Information）
        - 深度估计模型被广泛用于研究图像中物体的体积构成。这是计算机图形学领域的一个重要应用场景。
    - 3D重建（3D Representation）
        - 深度估计模型也可用于从2D图像构建3D表示。

- 提示：
    - 我们选择模型尽量选择Transformers实现的模型类型。

## 2.1. 任务模型的应用

### (1) DepthEstimationPipeline类说明

- DepthEstimationPipeline类的声明与集成结构
    - 类的声明：
        - `DepthEstimationPipeline(transformers.pipelines.base.Pipeline)`
    -  类的集成结构：
        -  DepthEstimationPipeline
        -  transformers.pipelines.base.Pipeline
        -  transformers.pipelines.base._ScikitCompat
        -  abc.ABC
        -  transformers.utils.hub.PushToHubMixin
        -  builtins.object

- DepthEstimationPipeline类的构造器说明（来自父类Pipeline）：
    - `def __init__(self, *args, **kwargs)`，其中可以使用的参数为（根据参数的声明方式，参数都必须是关键字参数，而且都可选）：
        - model参数：
            - 参数类型是PreTrainedModel，供pipeline用于预测的模型。
        - image_processor参数：
            - 参数类型是BaseImageProcessor，供pipeline用于为模型编码数据的图像处理器。
        - task参数：
            - 参数类型是str, 默认是""，pipeline的任务标识符
        - num_workers参数：
            - 参数类型是int, 默认是8，当pipeline使用DataLoader时（例如传入数据集，且在GPU上运行PyTorch模型），设置使用的工作进程数量。
        - batch_size参数：
            - 参数类型是int, 默认是1， 当pipeline使用DataLoader时（例如传入数据集，且在GPU上运行PyTorch模型），设置使用的批次大小。对于推理而言，增大批次大小并非总是有益，根据实际DataLoader大小需要设置，在推理的时候，建议取值1。大部分情况下，推理不一定使用DataLoader。
        - args_parser参数：
            - 参数类型是ArgumentHandler，负责解析提供的pipeline参数的对象的引用。主要提供参数的默认值以及定义参数，对推理而言一般采用默认。
        - device参数：
            - 参数类型是int, 默认是-1，负责设置CPU/GPU支持的设备序号。设置为-1将使用CPU，设置为正数将在对应的CUDA设备ID上运行模型。您也可以直接传入原生的torch.device对象或字符串。
        - dtype参数：
            - 参数类型是str或者torch.dtype, 用于为此模型指定可用的精度（例如 torch.float16, torch.bfloat16 … 或 "auto"）
        - binary_output参数：
            - 参数类型是bool, 默认是False，指示pipeline的输出应以序列化格式（例如pickle）还是原始输出数据（如文本）的形式进行。

- DepthEstimationPipeline类的核心函数：
    - inputs参数：指定需要推理的图像，图像格式可以为如下格式：
        - 包含图像http或者https链接的字符串
        - 包含图像本地路径的字符串
        - 直接通过PIL加载的图像
    - timeout参数：该参数可以单独使用，也可以使用关键字为parameters的字典参数。 该参数主要用来控制input图像下载的延时时间（超过这个时间值，则放弃下载）。
    - 返回值：
        - `dict[str, Any]`字典格式或者`list[dict[str, Any]]`字典列表（表示多张图像推理）
```python
def __call__(
    self, 
    inputs: Union[str, list[str], "Image.Image", list["Image.Image"]], 
    **kwargs: Any
    ) -> dict[str, Any] | list[dict[str, Any]]
```

- DepthEstimationPipeline的辅助处理器状态（）
    - `_load_processor = False`
    - **`_load_image_processor = True`**
    - `_load_feature_extractor = False`
    - `_load_tokenizer = False`

## 2.2. DepthEstimationPipeline的技术分析

### (1) 数据的预处理

- 数据的预处理是由DPTImageProcessor类负责。负责把不同格式的图像抽取特征，并转换为模型推理需要的输入格式（张量）,下面是对应的成员函数：
    - preprocess：预处理。

- DepthProImageProcessor类的构造器，构造器对应配置参数的加载（配置文件在模型文件中的文件名是：`preprocessor_config.json`）：
    - do_resize参数：是否将图像调整为指定的大小。参数类型是bool，默认设置为True。指定的大小由size参数指定。
    - size参数：当设置参数do_resize=True，可以使用size参数设置resize的大小。参数类型是字典，字典包含两个关键字：`height`,`width`。该参数会被preprocess函数的参数覆盖（重载overload）。
    - resample参数：调整大小时使用的重采样方法。使用Pillow图像库中的重采样标志。通常默认值是 3（即 BICUBIC）（image_utils模块，InterpolationMode来自torchvision.transforms）。
        - PILImageResampling.NEAREST: InterpolationMode.NEAREST_EXACT,
        - PILImageResampling.BOX: InterpolationMode.BOX,
        - PILImageResampling.BILINEAR: InterpolationMode.BILINEAR,
        - PILImageResampling.HAMMING: InterpolationMode.HAMMING,
        - PILImageResampling.BICUBIC: InterpolationMode.BICUBIC,
        - PILImageResampling.LANCZOS: InterpolationMode.LANCZOS,
    - keep_aspect_ratio参数：设置是否保持图像高宽比。默认是False。
    - ensure_multiple_of参数：如果do_resize=True，则使用该参数，确保size是该参数的倍数，该参数默认为1。
    - do_rescale参数：是否对像素值进行重新缩放。通常是将原始像素范围 `[0, 255]` 缩放到 `[0, 1]`。
    - rescale_factor参数：重新缩放的因子。通常为 1/255
    - do_normalize参数：是否对图像进行标准化。这通常是必要的，以便输入数据的分布与模型预训练时的分布一致。
    - image_mean参数：用于标准化图像的均值向量和标准差向量。这些值通常是 ImageNet 数据集的统计值
    - image_std参数：用于标准化图像的均值向量和标准差向量。这些值通常是 ImageNet 数据集的统计值
    - do_pad参数：是否对图像进行填充。对于 DPT 模型，有时需要将图像填充到能被特定倍数（如 patch 大小）整除的尺寸，以简化后续的分块处理。
    - size_divisor参数：如果 do_pad=True，图像尺寸将被填充为该参数的倍数。这通常与模型使用的 patch size 相关（例如，DPT 使用 ViT 时，patch size 可能是 16，size_divisor 可能设置为 16）。
    - do_reduce_labels参数：该参数在图像分割的时候使用，参数类型是bool，设置是否把分割图像的所有标签减1。通常用于背景值为0的数据集，且背景本身并不包含在数据集的所有类别中（例如ADE20k）。背景标签将被替换为255。
    - kwargs参数：其他参数。比如来自父类BaseImageProcessor的构造器参数。
    - 提示：上面参数无序过多关注，只需要了解其作用，并且了解模型文件提供的配置就行。

```python
 def __init__(
```

### (3) 推理后处理

- 从上面看见模型推理的结果实际已经是深度估计了，但是大小是按照预处理图像输出的，所以后处理的功能是：
    - 将深度图上采样回原始图像尺寸

````python
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from transformers import DPTImageProcessor, DPTForDepthEstimation

plt.rcParams["font.family"] = ["Microsoft YaHei"]
model_file = "F:/03Models/dpt-large"
img_file = "./imgs/scene.jpg"

# 1. 加载模型和处理器
processor = DPTImageProcessor.from_pretrained(model_file)
model = DPTForDepthEstimation.from_pretrained(model_file)

# 将模型设置为评估模式，并移到GPU（如果可用）
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()
print(f"模型已加载至: {device}")

# 2. 获取输入图像
image = Image.open(img_file)
print(f"原始图像尺寸: {image.size}")

# 3. 图像预处理并推理
# 使用processor将图像转换为模型所需的格式 [citation:7]
inputs = processor(images=image, return_tensors="pt").to(device)

with torch.no_grad():
    outputs = model(**inputs)
    # 获取原始深度预测张量 
    predicted_depth = outputs.predicted_depth

# 4. 后处理：将深度图上采样回原始图像尺寸 
# predicted_depth 的shape是 (1, H, W)，需要增加一个维度用于interpolate
prediction = torch.nn.functional.interpolate(
    predicted_depth.unsqueeze(1),  # 在第二维增加一个维度
    size=image.size[::-1],  # PIL image size 是 (width, height)，所以需要[::-1]
    mode="bicubic",   # 插值算法
    align_corners=False,
)

# 将深度张量转换为numpy数组
depth_map = prediction.squeeze().cpu().numpy()
print(f"深度图尺寸: {depth_map.shape}")

# 5. 可视化深度图
plt.figure(figsize=(12, 4))

# 显示原始图像
plt.subplot(131)
plt.imshow(image)
plt.title("原始图像")
plt.axis('off')

# 显示深度灰度图（值越接近1（亮）表示越近）
depth_normalized = (depth_map - depth_map.min()) / (depth_map.max() - depth_map.min())  # 标准化（归一化），可以使用imshow直接归一化
plt.subplot(132)
plt.imshow(depth_normalized, cmap='gray')
plt.title("深度图 (灰度)")
plt.axis('off')

# 显示深度彩色热力图
plt.subplot(133)
plt.imshow(depth_map, cmap='hot')
plt.title("深度图 (热力图)")
plt.axis('off')

plt.tight_layout()
plt.show()
````

### (4) DepthEstimationPipeline的完整实现流程

- 上面通过预处理，推理，后处理的流程说明了DepthEstimationPipeline的实现框架。下面是Transformers的完整实现类图。

- 下面是总体工作流程框架:
    - 

````python
import torch
import numpy as np
import matplotlib.pyplot as plt
from transformers import DepthEstimationPipeline, DPTImageProcessor, DPTForDepthEstimation

plt.rcParams["font.family"] = ["Microsoft YaHei"]

model_file = "F:/03Models/dpt-large"
img_file = "./imgs/scene.jpg"

processor = DPTImageProcessor.from_pretrained(model_file)
model = DPTForDepthEstimation.from_pretrained(model_file)
# Pipeline
pipe = DepthEstimationPipeline(
    model=model,
    image_processor=processor,
    task="depth-estimation",
    device="cuda",
)

# 预处理
inputs = pipe.preprocess(img_file)
# 推理
outputs = pipe.forward(inputs)
# 后处理
depth_estimation = pipe.postprocess(outputs)
print(depth_estimation)

result = depth_estimation["predicted_depth"]

# 显示深度灰度图（值越接近1（亮）表示越近）
depth_normalized = (result - result.min()) / (result.max() - result.min())  # 标准化（归一化），可以使用imshow直接归一化
plt.figure(figsize=(5, 4))
plt.imshow(depth_normalized, cmap='hot')
plt.axis('off')
plt.show()
````

- 代码说明：
    - 在preprocess函数中就是调用image_processor对象完成图像预处理。
    - 在forward函数中调用model对象完成推理。
    - 在postprocess函数中调用image_processor.post_process_depth_estimation完成后处理。

- preprocess函数的实现过程（官方源代码）：
```python
def preprocess(self, image, timeout=None):
    image = load_image(image, timeout)
    model_inputs = self.image_processor(images=image, return_tensors=self.framework)
    if self.framework == "pt":
        model_inputs = model_inputs.to(self.dtype)
    model_inputs["target_size"] = image.size[::-1]
    return model_inputs
```

- forward函数的实现过程（官方源代码）:

```python
def forward(self, model_inputs, **forward_params):
    # ...省略无关代码
    model_outputs = self._forward(model_inputs, **forward_params)
    # ...省略无关代码
    return model_outputs
def _forward(self, model_inputs):
    target_size = model_inputs.pop("target_size")
    model_outputs = self.model(**model_inputs)
    model_outputs["target_size"] = target_size
    return model_outputs
```

- postprocess函数的实现过程（官方代码）

```python
def postprocess(self, model_outputs):
    outputs = self.image_processor.post_process_depth_estimation(
        model_outputs,
        # this acts as `source_sizes` for ZoeDepth and as `target_sizes` for the rest of the models so do *not*
        # replace with `target_sizes = [model_outputs["target_size"]]`
        [model_outputs["target_size"]],
    )

    formatted_outputs = []
    for output in outputs:
        depth = output["predicted_depth"].detach().cpu().numpy()
        depth = (depth - depth.min()) / (depth.max() - depth.min())
        depth = Image.fromarray((depth * 255).astype("uint8"))

        formatted_outputs.append({"predicted_depth": output["predicted_depth"], "depth": depth})

    return formatted_outputs[0] if len(outputs) == 1 else formatted_outputs
```
